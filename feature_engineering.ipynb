```python
from nltk.corpus import stopwords
from nltk.corpus.reader import WordListCorpusReader

stp_words = stopwords.words('arabic')
more_stp_words = WordListCorpusReader('.', ['/workspaces/Capstone-Project/arabic.txt'])

stp_words.extend(more_stp_words.words())
stp_words = set(stp_words)
print(len(stp_words))
```

```python
news['content_without_stopwords'] = news['content'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stp_words)]))
```

```python
from camel_tools.utils.dediac import dediac_ar

# apply to your text column
news.content_without_stopwords = news.content_without_stopwords.apply(dediac_ar)
```

```python
from camel_tools.utils.normalize import normalize_alef_maksura_ar
from camel_tools.utils.normalize import normalize_alef_ar
from camel_tools.utils.normalize import normalize_teh_marbuta_ar

def normalize(sentence):
    sentence = normalize_alef_maksura_ar(sentence)
    sentence = normalize_alef_ar(sentence)
    sentence = normalize_teh_marbuta_ar(sentence)
    return sentence
  
news.content_without_stopwords = news.content_without_stopwords.apply(normalize)
```

```python
from nltk import sent_tokenize

news["content_tokenized"] = news.content_without_stopwords.apply(sent_tokenize)
news["title_tokenized"] = news.title.apply(sent_tokenize)
news["author_tokenized"] = news.author.apply(sent_tokenize)
```

```python
import heapq
from nltk import word_tokenize
def get_summary(content, tokenized_content):
	word_frequency = {}
	for word in word_tokenize(content):
		if word not in word_frequency.keys():
			word_frequency[word] = 1
		else:
			word_frequency[word] += 1
			
	maximum_frequncy = max(word_frequency.values())
	for word in word_frequency.keys():
		word_frequency[word] = (word_frequency[word]/maximum_frequncy)
	
	sentence_scores = {}
	for sent in tokenized_content:
		for word in word_tokenize(sent.lower()):
			if word in word_frequency.keys():
				if len(sent.split(' ')) < 30:
					if sent not in sentence_scores.keys():
						sentence_scores[sent] = word_frequency[word]
					else:
						sentence_scores[sent] += word_frequency[word]
	
	summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

	summary = ' '.join(summary_sentences)
	return summary
```

```python
summary = []
for i, row in news.iterrows():
	summary.append(get_summary(row["content_without_stopwords"], row["content_tokenized"]))
```
